## Tokenization using Regular Expression

```python
import re
Re_tokens = re.findall(r'\b[\w-]+\b', text)
print("There are ", len(Re_tokens), "tokens in this text\n")

counter = 0
for w in Re_tokens:
    counter+=1
    print(counter,".",w)
    
print("\nText List ~~~\n",Re_tokens)
```

## Tokenization using NLTK

```python
import nltk
from nltk.tokenize import word_tokenize
nltk_tokens = word_tokenize(text)
print("There are ", len(nltk_tokens), "tokens in this text\n")
print(nltk_tokens)


counter = 0
for w in nltk_tokens:
    counter+=1
    print(counter,".",w)

# Use nltk.Text() to create a text list from the tokens list.
Tokenstext = nltk.Text(nltk_tokens)
print("\nText List ~~~\n", Tokenstext[0:len(Tokenstext)])
```

## Tokenization using textblob

```python
from textblob import TextBlob
textblob_tokens = TextBlob(text).words
print("There are ", len(textblob_tokens), "tokens in this text\n")

counter = 0
for w in textblob_tokens:
    counter+=1
    print(counter,".",w)
    
print("\nText List ~~~\n",textblob_tokens)

```

### Removing Stop Words & Punctuations

```python
import string

text_lower = text.lower()
wordtokens = nltk.tokenize.word_tokenize(text_lower)
print("Tokens ~~~ \n", wordtokens)

print("\nTotal number of words in this text corpus is",len(wordtokens))

stopTokens = nltk.corpus.stopwords.words("english") + list(string.punctuation)
filteredTokens = []

for w in wordtokens:
    if w not in stopTokens :
        filteredTokens.append(w)

print("There are", len(filteredTokens), "words in this text after removing stop words\n")
print("Filtered Tokens ~~~ \n", filteredTokens)

print("\nStop Tokens ~~~ \n", stopTokens)

fdist1 = nltk.FreqDist(filteredTokens)
print("\nFrequency Distribution ~~~\n",fdist1.most_common(10))
```
